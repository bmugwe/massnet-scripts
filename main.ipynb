```python
import pandas as pd
from sqlalchemy import create_engine, text
from urllib.parse import quote_plus

password = "123456@Ab"
encoded_password = quote_plus(password)

# Create the URL
db_url = "mysql+mysqlconnector://{USER}:{PWD}@{HOST}/{DBNAME}"

# MySQL connection strings
source_db_url = db_url.format(
    USER="root", PWD=encoded_password, HOST="localhost:3306", DBNAME="massnet_prod"
)

# target_db_url = f"mysql://root:{encoded_password}@localhost:3306/massnet"  # Replace with your target DB credentials
target_db_url = db_url.format(
    USER="root", PWD=encoded_password, HOST="localhost:3306", DBNAME="massnet"
)
# Table names
table_name = "master_lists"  # Replace with your table name

# Create SQLAlchemy engines for source and target databases
source_engine = create_engine(source_db_url, echo=False)
target_engine = create_engine(target_db_url)
# 
# Fetch data from source database
query = f"SELECT * FROM {table_name} limit 1"
try:
    df = pd.read_sql(query, source_engine)
except Exception as e:
    print(f"Error reading:  {e}")

# print(f"Data: {df}")

# Push data to target database using INSERT IGNORE
insert_query = f"INSERT INTO {table_name} VALUES %s"

#Convert DataFrame to a list of tuples
data_tuples = [tuple(row) for row in df.values]
print('--------------')
print(data_tuples)

# Execute the REPLACE / INSERT IGNORE statement using SQLAlchemy's execute method
with target_engine.connect() as connection:
    for row in df.values:
        tuple_row = row
        tuple_row[tuple_row==None] = 'null'
        tuple_rowd = [tuple(tuple_row)]
        print(type(tuple_rowd), tuple_rowd)
    execution_string  = (text(insert_query), tuple_rowd)
    print(execution_string)

    connection.execute(text(insert_query),tuple_rowd)
    print("Data inserted into target database with IGNORE.")
try:
    pass
except Exception as e:
    print(f"Error inserting data: {e}")
```
